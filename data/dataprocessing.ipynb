```python
# Loại bỏ các dòng có giá trị âm trong cột 'Charging Time'
df = df[df['Average Power'] >= 0]

df = df[df['Total Energy Delivered']>=0]
```

```python
df = df[~(df['Charging Time'].isna() & df['SOC Charged'].isna())]
```

```python
print(df.isnull().sum())
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

train_data = df.dropna(subset=['Charging Time', 'SOC Charged'])

X_train = train_data[['Average Power', 'Total Energy Delivered', 'SOC Charged']]
y_train = train_data['Charging Time']
X_missing = df[df['Charging Time'].isna()][['Average Power', 'Total Energy Delivered', 'SOC Charged']]

model = LinearRegression()

model.fit(X_train, y_train)


predicted_values = model.predict(X_missing)
predicted_values = np.clip(predicted_values, a_min=0, a_max=None)

df.loc[df['Charging Time'].isna(), 'Charging Time'] = predicted_values
print(df['Charging Time'].isna().sum())
```

```python
from sklearn.linear_model import LinearRegression

train_data_soc = df.dropna(subset=['SOC Charged'])

X_train_soc = train_data_soc[['Charging Time', 'Average Power', 'Total Energy Delivered']]
y_train_soc = train_data_soc['SOC Charged']

X_missing_soc = df[df['SOC Charged'].isna()][['Charging Time', 'Average Power', 'Total Energy Delivered']]

model_soc = LinearRegression()

model_soc.fit(X_train_soc, y_train_soc)
predicted_soc_values = model_soc.predict(X_missing_soc)

df.loc[df['SOC Charged'].isna(), 'SOC Charged'] = predicted_soc_values

print(df['SOC Charged'].isna().sum())
```
